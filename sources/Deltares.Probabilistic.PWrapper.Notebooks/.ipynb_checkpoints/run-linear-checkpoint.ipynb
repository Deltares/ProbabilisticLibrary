```python
from ptk import *

stochast = Stochast()
stochast.distribution = 'uniform'
stochast.minimum = -1
stochast.maximum = 1

print('mean = ' + str(stochast.mean))
print('sigma = ' + str(stochast.deviation))
```

```python
def linear(a, b):
    return 1.9 - (a+b)
```

```python
project = ReliabilityProject()

project.model = linear

stochast1 = project.get_variable('a')
stochast1.distribution = 'uniform'
stochast1.minimum = -1
stochast1.maximum = 1

stochast2 = project.get_variable('b')
stochast2.distribution = 'uniform'
stochast2.minimum = -1
stochast2.maximum = 1

project.settings.reliability_method = 'form'

project.run()

dp = project.design_point
beta = dp.reliability_index

print('beta = ' + str(beta))

```

```python
for alpha in dp.alphas:
    print(alpha.variable.name + ': alpha = ' + str(alpha.alpha) + ', x = ' + str(alpha.x))
```

```python
project.correlation_matrix.set_correlation (stochast1, stochast2, 0.8)

project.run()

dp = project.design_point
beta = dp.reliability_index

print('beta = ' + str(beta))

for alpha in dp.alphas:
    print(alpha.variable.name + ': alpha = ' + str(alpha.alpha) + ', x = ' + str(alpha.x))
```
