```python
from ptk import *
import numpy as np

def hunt(Tp, tanA, Hs, hCrest, MWL):
    g = 9.8
    L0 = g  * Tp * Tp
    ksi = tanA / np.sqrt(2 * np.pi * Hs / L0)
    Ru = ksi * Hs

    z = hCrest - (MWL + Ru)
    return z

project = ReliabilityProject()
project.model = hunt

[print(v) for v in project.variables]
```

```python
project.variables['Tp'].distribution = 'normal'
project.variables['Tp'].mean = 6
project.variables['Tp'].deviation = 2

project.variables['tanA'].distribution = 'deterministic'
project.variables['tanA'].mean = 0.333333

project.variables['Hs'].distribution = 'normal'
project.variables['Hs'].mean = 3
project.variables['Hs'].deviation = 1

project.variables['hCrest'].distribution = 'deterministic'
project.variables['hCrest'].mean = 10

project.variables['MWL'].distribution = 'exponential'
project.variables['MWL'].shift = 0.5
project.variables['MWL'].scale = 1

```
