# Distribution functions
In this example, we will demonstrate how to define a distribution function using the probabilistic library.

The library supports the following distribution functions: `deterministic`, `normal`, `log_normal`, `uniform`, `triangular`, `trapezoidal`, `exponential`, `gumbel`, `weibull`, `conditional_weibull`, `frechet`, `generalized_extreme_value`, `rayleigh`, `rayleigh_n`, `pareto`, `generalized_pareto`, `student_t`, `gamma`, `beta`, `poisson`, `discrete`, `bernoulli`, `cdf_curve`, `histogram`.

### Define a stochastic variable
First, we import the necessary package:

```python
from probabilistic_library import Stochast, DistributionType, DiscreteValue, HistogramValue, FragilityValue
import numpy as np
```

Next, we create a stochastic variable object:

```python
stochast = Stochast()
```

We use the `stochast.get_pdf()` and `stochast.get_cdf()` methods to plot the probability density function and the cumulative distribution function, respectively. 

```python
from utils.plotting import plot_dist
```

Below, we demonstrate how to define distribution functions in the library.

### Deterministic distribution

```python
stochast.distribution = DistributionType.deterministic
stochast.mean = 0.8

stochast.print()
```

### Normal distribution

```python
stochast = Stochast()
stochast.distribution = DistributionType.normal
stochast.location = 2.0
stochast.scale = 0.5

stochast.print()
stochast.plot()

val_grid = np.arange(stochast.location-2.0, stochast.location+2.0, 0.01)
plot_dist(val_grid, stochast)
```

### Log-normal distribution

```python
stochast = Stochast()
stochast.distribution = DistributionType.log_normal
stochast.location = 0.0
stochast.scale = 1.0
stochast.shift = 0.0

stochast.print()
stochast.plot()

val_grid = np.arange(stochast.shift, 10.0, 0.01)
plot_dist(val_grid, stochast)
```

### Uniform distribution

```python
stochast = Stochast()
stochast.distribution = DistributionType.uniform
stochast.minimum = -1
stochast.maximum = 1

stochast.print()
stochast.plot()

val_grid = np.arange(-2.0, 2.0, 0.01)
plot_dist(val_grid, stochast)
```

### Triangular distribution

```python
stochast = Stochast()
stochast.distribution = DistributionType.triangular
stochast.minimum = 0.0 # parameter a
stochast.maximum = 3.0 # parameter b
stochast.shift = 2.0 # parameter c

stochast.print()
stochast.plot()

val_grid = np.arange(stochast.minimum, stochast.maximum, 0.01)
plot_dist(val_grid, stochast)
```

### Trapezoidal distribution

```python
stochast = Stochast()
stochast.distribution = DistributionType.trapezoidal
stochast.minimum = 1.0 # parameter a
stochast.maximum = 4.5 # parameter b
stochast.shift = 2.0 # parameter c
stochast.shift_b = 3.0 # parameter d

val_grid = np.arange(stochast.minimum, stochast.maximum, 0.01)
plot_dist(val_grid, stochast)
```

### Exponential distribution

```python
stochast = Stochast()
stochast.distribution = DistributionType.exponential
stochast.scale = 2.0
stochast.shift = 0.5

val_grid = np.arange(stochast.shift, 10.0, 0.01)
plot_dist(val_grid, stochast)
```

### Gumbel distribution

```python
stochast = Stochast()
stochast.distribution = DistributionType.gumbel
stochast.scale = 2.0
stochast.shift = 3.0

val_grid = np.arange(-5.0, 15.0, 0.01)
plot_dist(val_grid, stochast)
```

### Weibull distribution

```python
stochast = Stochast()
stochast.distribution = DistributionType.weibull
stochast.scale = 1.0
stochast.shift = 1.0
stochast.shape = 2.0

val_grid = np.arange(stochast.shift, 8.0, 0.01)
plot_dist(val_grid, stochast)
```

### Conditional Weibull distribution

```python
stochast = Stochast()
stochast.distribution = DistributionType.conditional_weibull
stochast.scale = 1.0
stochast.shift = 5.0
stochast.shape = 2.0
stochast.shape_b = 1.0

val_grid = np.arange(4.5, 8.0, 0.01)
plot_dist(val_grid, stochast)
```

### Frechet distribution

```python
stochast = Stochast()
stochast.distribution = DistributionType.frechet
stochast.scale = 0.5
stochast.shift = 1.0
stochast.shape = 2.0

val_grid = np.arange(stochast.shift, 5.0, 0.01)
plot_dist(val_grid, stochast)
```

### Generalized extreme value distribution

```python
stochast = Stochast()
stochast.distribution = DistributionType.generalized_extreme_value
stochast.scale = 1.0
stochast.shift = 5.0
stochast.shape = -0.5

if stochast.shape<0.0: # grid when the shape parameter is negative
    limit = stochast.shift - stochast.scale/stochast.shape
    val_grid = np.arange(1.0, limit, 0.01)
elif stochast.shape>0.0: # grid when the shape parameter is positive
    limit = stochast.shift - stochast.scale/stochast.shape
    val_grid = np.arange(limit, 10.0, 0.01)
else:
    val_grid = np.arange(-2.0, 10.0, 0.01)

plot_dist(val_grid, stochast)
```

### Rayleigh distribution

```python
stochast = Stochast()
stochast.distribution = DistributionType.rayleigh
stochast.scale = 0.5
stochast.shift = 1.0

val_grid = np.arange(stochast.shift, 4.0, 0.01)
plot_dist(val_grid, stochast)
```

### Rayleigh N distribution

### Generalized Pareto distribution

### Gamma distribution

```python
stochast = Stochast()
stochast.distribution = DistributionType.gamma
stochast.scale = 2.0
stochast.shape = 4.0

val_grid = np.arange(0.0, 25.0, 0.01)
plot_dist(val_grid, stochast)
```

### Beta distribution

```python
stochast = Stochast()
stochast.distribution = DistributionType.beta
stochast.shape = 2.0
stochast.shape_b = 2.0

val_grid = np.arange(0.0, 1.0, 0.01)
plot_dist(val_grid, stochast)
```

### Poisson distribution

```python
stochast = Stochast()
stochast.distribution = DistributionType.poisson
stochast.location = 4.0

stochast.print()
stochast.plot()

val_grid = np.arange(0.0, 20.0, 0.01)
plot_dist(val_grid, stochast)
```

### Discrete distribution

```python
stochast = Stochast()
stochast.distribution = DistributionType.discrete

x = [1, 2, 3]
amount = [0.1, 0.4, 0.5]

for i in range(0, len(x)):
    
    dv = DiscreteValue()
    dv.x = x[i]
    dv.amount = amount[i]

    stochast.discrete_values.append(dv)

stochast.print()
stochast.plot()

val_grid = np.arange(0.0, 4.0, 0.01)
plot_dist(val_grid, stochast)
```

### Bernoulli distribution

```python
stochast = Stochast()
stochast.distribution = DistributionType.bernoulli
stochast.mean = 0.3

stochast.print()
stochast.plot()

val_grid = np.arange(-1.0, 2.0, 0.01)
plot_dist(val_grid, stochast)
```

### CDF curve distribution

```python
stochast = Stochast()
stochast.distribution =  DistributionType.cdf_curve

x = [1, 2, 3]
probability_of_failure = [0.01, 0.001, 0.0001] 

# these are the exceedance probabilities, which will shift the probability mass to an area < 1.0

for i in range(0, len(x)):

    fc = FragilityValue()
    fc.x = x[i]
    fc.probability_of_failure = probability_of_failure[i]
    
    stochast.fragility_values.append(fc)

stochast.print()
stochast.plot()

val_grid = np.arange(-8.0, 4.0, 0.01)
plot_dist(val_grid, stochast)
```

### Histogram distribution

```python
stochast = Stochast()
stochast.distribution =  DistributionType.histogram

lower_bound = [1, 2, 3]
upper_bound = [2, 3, 4]
amount = [100, 200, 300]

for i in range(0, len(lower_bound)):

    hv = HistogramValue()
    hv.lower_bound = lower_bound[i]
    hv.upper_bound = upper_bound[i]
    hv.amount = amount[i]
    stochast.histogram_values.append(hv)

stochast.print()
stochast.plot()

val_grid = np.arange(0.0, 4.0, 0.01)
plot_dist(val_grid, stochast)
```
